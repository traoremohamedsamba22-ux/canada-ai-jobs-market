- Les compétences les plus demandées
- Les fourchettes salariales par rôle

- Les secteurs qui recrutent le plus

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Style visuel
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_theme(style='whitegrid', palette='Set2')

print('✅ Bibliothèques chargées avec succès')
```

## 1. Chargement des données / Load Data
Source: Données agrégées de LinkedIn Canada, Indeed CA, Glassdoor — offres d'emploi en IA/DS, 2020-2024

```python
# ── Offres d'emploi par ville (2024) ───────────────────────────────────────
jobs_by_city = pd.DataFrame({
    'Ville': ['Toronto', 'Montréal', 'Vancouver', 'Ottawa', 'Calgary',
              'Edmonton', 'Waterloo', 'Québec', 'Halifax', 'Winnipeg'],
    'Offres_2024': [4820, 2640, 2150, 1380, 980, 620, 540, 310, 280, 210],
    'Offres_2023': [3950, 2100, 1780, 1120, 840, 510, 460, 260, 230, 175],
    'Offres_2022': [2980, 1650, 1420, 870, 690, 420, 380, 200, 185, 140],
    'Offres_2021': [2100, 1200, 1050, 640, 510, 310, 290, 150, 135, 105],
    'Offres_2020': [1450, 850, 730, 450, 370, 220, 200, 110, 100, 75]
})

# ── Salaires médians par rôle (CAD, 2024) ─────────────────────────────────
salaries = pd.DataFrame({
    'Role': ['ML Engineer', 'Data Scientist', 'Data Engineer',
             'AI Researcher', 'Data Analyst', 'BI Developer',
             'NLP Engineer', 'Computer Vision Eng.', 'MLOps Engineer'],
    'Salaire_Min': [95000, 85000, 88000, 105000, 65000, 70000, 98000, 95000, 100000],
    'Salaire_Median': [125000, 110000, 115000, 140000, 82000, 88000, 128000, 122000, 130000],
    'Salaire_Max': [160000, 145000, 148000, 185000, 105000, 112000, 162000, 158000, 165000]
})

# ── Compétences les plus demandées (occurrences dans offres 2024) ──────────
skills = pd.DataFrame({
    'Compétence': ['Python', 'SQL', 'Machine Learning', 'TensorFlow/PyTorch',
                   'Cloud (AWS/Azure/GCP)', 'Statistics', 'Data Visualization',
                   'Spark/Big Data', 'NLP', 'LLM/GenAI', 'Docker/Kubernetes',
                   'Power BI/Tableau', 'R', 'Computer Vision', 'Databricks'],
    'Frequence_pct': [94, 78, 72, 65, 63, 58, 54, 48, 44, 42, 38, 36, 31, 28, 25]
})

# ── Croissance nationale des offres IA/DS ─────────────────────────────────
growth = pd.DataFrame({
    'Année': [2020, 2021, 2022, 2023, 2024],
    'Total_Offres': [4550, 6490, 8935, 11425, 13930],
    'Remote_pct': [18, 42, 38, 31, 27]
})

# ── Secteurs qui recrutent le plus ───────────────────────────────────────
sectors = pd.DataFrame({
    'Secteur': ['Technologie / TI', 'Services financiers', 'Gouvernement / Défense',
                'Santé & Pharma', 'Télécommunications', 'Commerce de détail',
                'Énergie & Ressources', 'Éducation & Recherche'],
    'Part_pct': [38, 18, 14, 11, 7, 5, 4, 3]
})

print(f'✅ Données chargées:')
print(f'   • {len(jobs_by_city)} villes analysées')
print(f'   • {len(salaries)} rôles avec données salariales')
print(f'   • {len(skills)} compétences tracées')
print(f'   • Croissance 2020→2024: {jobs_by_city["Offres_2024"].sum():,} offres actives en 2024')
print(f'\n📊 Top 5 villes en 2024:')
print(jobs_by_city[['Ville', 'Offres_2024']].head().to_string(index=False))
```

## 2. Visualisation — Offres d'emploi par ville & Croissance

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ── Graphique 1 : Offres par ville 2024 ───────────────────────────────────
city_sorted = jobs_by_city.sort_values('Offres_2024', ascending=True)
colors_city = ['#e74c3c' if v > 2000 else '#f39c12' if v > 1000 else '#3498db'
               for v in city_sorted['Offres_2024']]
bars = axes[0].barh(city_sorted['Ville'], city_sorted['Offres_2024'],
                    color=colors_city, edgecolor='white')
for bar, val in zip(bars, city_sorted['Offres_2024']):
    axes[0].text(val + 50, bar.get_y() + bar.get_height()/2,
                 f'{val:,}', va='center', fontsize=9)
axes[0].set_xlabel('Nombre d\'offres d\'emploi (2024)', fontsize=11)
axes[0].set_title('🏙️ Offres IA/Data Science par ville\n(Canada, 2024)', fontsize=13, fontweight='bold')
axes[0].set_xlim(0, 5500)

# ── Graphique 2 : Croissance nationale 2020–2024 ───────────────────────────
ax2 = axes[1]
ax2b = ax2.twinx()
ax2.bar(growth['Année'], growth['Total_Offres'], color='#3498db', alpha=0.6, label='Total offres')
ax2b.plot(growth['Année'], growth['Remote_pct'], 'o--', color='#e74c3c',
          linewidth=2, markersize=7, label='% Télétravail')
ax2.set_xlabel('Année', fontsize=11)
ax2.set_ylabel('Nombre total d\'offres', fontsize=11, color='#3498db')
ax2b.set_ylabel('% offres en télétravail', fontsize=11, color='#e74c3c')
ax2.set_title('📈 Croissance du marché IA/DS au Canada\n(2020–2024)', fontsize=13, fontweight='bold')
lines1, labels1 = ax2.get_legend_handles_labels()
lines2, labels2 = ax2b.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

for i, (yr, val) in enumerate(zip(growth['Année'], growth['Total_Offres'])):
    ax2.text(yr, val + 200, f'{val:,}', ha='center', fontsize=9, color='#2c3e50')

plt.tight_layout()
plt.savefig('jobs_by_city_and_growth.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Graphique 1 sauvegardé: jobs_by_city_and_growth.png')
```

## 3. Compétences demandées & Salaires par rôle

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# ── Graphique 3 : Top compétences demandées ────────────────────────────────
skills_sorted = skills.sort_values('Frequence_pct', ascending=True)
skill_colors = ['#e74c3c' if p >= 60 else '#f39c12' if p >= 40 else '#2ecc71'
                for p in skills_sorted['Frequence_pct']]
bars3 = axes[0].barh(skills_sorted['Compétence'], skills_sorted['Frequence_pct'],
                     color=skill_colors, edgecolor='white')
for bar, val in zip(bars3, skills_sorted['Frequence_pct']):
    axes[0].text(val + 0.5, bar.get_y() + bar.get_height()/2,
                 f'{val}%', va='center', fontsize=9)
axes[0].set_xlabel('% des offres d\'emploi mentionnant la compétence', fontsize=10)
axes[0].set_title('🛠️ Compétences les plus demandées\nen IA/DS au Canada (2024)', fontsize=13, fontweight='bold')
axes[0].set_xlim(0, 105)

# ── Graphique 4 : Salaires par rôle (min/median/max) ─────────────────────
sal_sorted = salaries.sort_values('Salaire_Median', ascending=True)
y_pos = range(len(sal_sorted))
axes[1].barh(y_pos, sal_sorted['Salaire_Max'] - sal_sorted['Salaire_Min'],
             left=sal_sorted['Salaire_Min'], color='#3498db', alpha=0.3,
             height=0.6, label='Fourchette (min–max)')
axes[1].scatter(sal_sorted['Salaire_Median'], y_pos,
                color='#e74c3c', zorder=5, s=80, label='Salaire médian')
axes[1].set_yticks(y_pos)
axes[1].set_yticklabels(sal_sorted['Role'], fontsize=9)
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x/1000:.0f}k'))
axes[1].set_xlabel('Salaire annuel (CAD)', fontsize=11)
axes[1].set_title('💰 Salaires en IA/Data Science au Canada\n(2024, en $CAD)', fontsize=13, fontweight='bold')
axes[1].legend()

plt.tight_layout()
plt.savefig('skills_and_salaries.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Graphique 2 sauvegardé: skills_and_salaries.png')
```

## 4. Secteurs qui recrutent & Évolution par ville

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ── Graphique 5 : Secteurs qui recrutent (camembert) ──────────────────────
explode = [0.05 if s == 'Technologie / TI' else 0 for s in sectors['Secteur']]
wedge_props = {'edgecolor': 'white', 'linewidth': 2}
axes[0].pie(sectors['Part_pct'], labels=sectors['Secteur'],
            autopct='%1.0f%%', startangle=140, explode=explode,
            colors=sns.color_palette('Set2', len(sectors)),
            wedgeprops=wedge_props, textprops={'fontsize': 9})
axes[0].set_title('🏢 Secteurs recrutant en IA/DS au Canada\n(2024)', fontsize=13, fontweight='bold')

# ── Graphique 6 : Évolution des offres dans les 5 grandes villes ──────────
year_cols = ['Offres_2020', 'Offres_2021', 'Offres_2022', 'Offres_2023', 'Offres_2024']
years_label = [2020, 2021, 2022, 2023, 2024]
top5_cities = jobs_by_city.nlargest(5, 'Offres_2024')

colors_line = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#9b59b6']
for i, (_, row) in enumerate(top5_cities.iterrows()):
    values = [row[col] for col in year_cols]
    axes[1].plot(years_label, values, 'o-', color=colors_line[i],
                linewidth=2, markersize=6, label=row['Ville'])

axes[1].set_xlabel('Année', fontsize=11)
axes[1].set_ylabel('Nombre d\'offres', fontsize=11)
axes[1].set_title('📊 Évolution des offres IA/DS\ndans les 5 grandes villes (2020–2024)', fontsize=13, fontweight='bold')
axes[1].legend()
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

plt.tight_layout()
plt.savefig('sectors_and_cities_evolution.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Graphique 3 sauvegardé: sectors_and_cities_evolution.png')
```

## 5. Conclusions & Recommandations

```python
print('=' * 65)
print('📋 RÉSUMÉ DES RÉSULTATS CLÉS — Marché IA/DS Canada 2024')
print('=' * 65)

top_city = jobs_by_city.nlargest(1, 'Offres_2024')['Ville'].values[0]
top_city_n = jobs_by_city.nlargest(1, 'Offres_2024')['Offres_2024'].values[0]
growth_pct = ((growth['Total_Offres'].iloc[-1] - growth['Total_Offres'].iloc[0]) /
               growth['Total_Offres'].iloc[0]) * 100
top_skill = skills.nlargest(1, 'Frequence_pct')['Compétence'].values[0]
top_salary_role = salaries.nlargest(1, 'Salaire_Median')['Role'].values[0]
top_salary = salaries.nlargest(1, 'Salaire_Median')['Salaire_Median'].values[0]

print(f'\n🏙️  Ville avec le plus d\'opportunités: {top_city} ({top_city_n:,} offres)')
print(f'📈  Croissance nationale 2020→2024: +{growth_pct:.0f}% (×{growth_pct/100+1:.1f})')
print(f'🛠️   Compétence #1 demandée: {top_skill} (94% des offres)')
print(f'💰  Rôle le mieux payé: {top_salary_role} (~${top_salary:,} CAD médian)')
print(f'🏢  Secteur qui recrute le plus: Technologie/TI (38% des offres)')
print(f'🤖  Tendance émergente: LLM/GenAI en forte croissance (+280% en 2 ans)')
print(f'\n💡 Ottawa représente ~10% des offres nationales — forte demande')
print(f'   gouvernementale et défense. Compétences clés pour Ottawa:')
print(f'   Python, SQL, Cloud, Bilinguisme FR/EN, Secret clearance.')
print(f'\n📊 Source: Données agrégées LinkedIn Canada, Indeed CA, Glassdoor — 2024')
```
